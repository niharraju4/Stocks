```python
import yfinance as yf
import pandas as pd

# Specify the stock symbol and time period
stock_symbol = "AAPL"  # Apple Inc. is used as an example
start_date = "2015-01-01"
end_date = "2024-02-11"

# Fetch historical data using yfinance
data = yf.download(stock_symbol, start=start_date, end=end_date, interval='1d')

# Display the fetched data
print(data.head())

# Save the data to a CSV file (optional)
data.to_csv(f"{stock_symbol}_30min_data.csv")

# Note: The 'interval' parameter in yfinance is used to specify the time interval (e.g., '1d' for daily, '1wk' for weekly, '1mo' for monthly, '30m' for 30 minutes, etc.)
```
